### алгоритм Кнута-Морриса-Пратта

Префикс-функция $\pi$ - это функция, которая строится для паттерна $P$ следующим образом:
$$ \pi[i] = \max \{k : P[0 \dots k-1] = P[i-k+1 \dots i] \} $$

Обновление:

При несовпадении $T[i] \neq P[j]$ индекс паттерна обновляется за $O(1)$:
$$
j \leftarrow \pi[j-1]
$$

Сложность:
* $O(N + M)$ по времени.
* $O(M)$ по памяти (в отличие от рабина карпа, где $O(1)$).
