### rolling hash

кольцевой хэш - это вид хэша, который для близких входов позволяет считать его за $O(1)$ вместо полного пересчета за $O(len(subseq))$.

Формула для подпоследовательности от i до i + m - 1 выглядит так:

$$
hash_i = (t_i * \alpha^{m-1} + ... + t_{i+m-1} \alpha^0) \mod p
$$

t_i - это числовое представление символа на i месте строки t

Тогда для подсчета хэша близкой к исходной подпоследовательности можно использовать уже рассчитанное значение hash_i, запишем формулу в быстром для рассчетов виде:

$$
hash_{i+1} = ((hash_i - t_i \alpha^{i-1}) \alpha + t_{i+m}) \mod p \equiv (t_{i+1} * \alpha^{m-1} + ... + t_{i+m} \alpha^0) \mod p
$$


### Сложность алгоритма

Сложность по времени:
- Лучший случай: $O(N + M)$. Предварительный подсчет хеша паттерна занимает $O(M)$. Проход по тексту занимает $O(N)$, так как пересчет хеша выполняется за $O(1)$.
- Худший случай: $O(NM)$. Происходит при очень частых коллизиях хеша (например, если текст и паттерн специфичны: "AAAAAAAAAAAAA", паттерн "AAAAB"). Тогда алгоритм на каждом шаге будет проверять посимвольно.
